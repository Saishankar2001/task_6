# Players with more runs in the League

```python
top_batsman = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
top_batsman
```

```python
plt.figure(figsize=(12,6))
top_batsman.head(10).plot(kind = 'bar',color = 'green')
plt.title("Players with more runs" , fontsize = 20)
plt.xlabel('batsman',fontsize = 20)
plt.ylabel('runs',fontsize = 20)
```

* Virat Kohli is the leading run scorer of the IPL.He has maintained the consistent performance in all the seasons..
* Most of them are the openers and experienced players of the team.
* We can also observe that half of them are the captains.

# Batsman with most number of Sixes..

```python
sixes = deliveries[deliveries['batsman_runs'] ==  6]
most_sixes = sixes.groupby('batsman').count()['batsman_runs'].sort_values(ascending = False)
```

```python
most_sixes = most_sixes.head(10)
most_sixes
```

```python
plt.figure(figsize=(12,6))
most_sixes.plot(kind = 'bar')
plt.show()
```

```python
top_bowler = deliveries.groupby('bowler')['player_dismissed'].count().sort_values(ascending = False)
top_bowler
```

```python
plt.figure(figsize=(12,6))
top_bowler.head(10).plot(kind = 'bar',color = 'red')
plt.title('Top wicket takers',fontsize = 20)
plt.xlabel("Bowlers", size = 15)
plt.ylabel("Total Wickets taken", size = 15)
```

* Lasith Malinga is the leading wicket taker followed by DJ Bravo,A Mishra..
* We can observe both fast bowlers and spinners.

# Winner of  IPL Seasons..

```python
winningteams = matches[['season','winner']]  ## A dataframe with seasons and winner
winningteams
```

```python
win_team = {}                              ## Created a dictionary to get winners of every season
for i in sorted(winningteams.season.unique()):
    win_team[i] = winningteams[winningteams.season == i]['winner'].tail(1).values[0]
    
winner_of_IPL = pd.Series(win_team)
winner_of_IPL = pd.DataFrame(winner_of_IPL, columns=['team'])
```

```python
winner_of_IPL
```

```python
winner_of_IPL['team'].value_counts().plot(kind = 'pie', figsize = (20,9), fontsize=12, autopct= '%1.1f%%', 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Teams", size = 10)
plt.xlabel("Frequency", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
winner_of_IPL['team'].value_counts().plot(kind = 'bar', figsize = (15,5), color = 'green')
plt.title("Winners of IPL across 12 seasons",fontsize=15,fontweight="bold")
plt.xlabel("Teams", size = 20)
plt.ylabel("no of times", size = 20)
plt.xticks(size = 10)
plt.yticks(size = 10)
```
